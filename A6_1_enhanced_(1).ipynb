##  Cell 4 | Step 4 — Outlier Detection & Handling

````python
#  PHASE 1 | STEP 4a : Detect outliers via IQR
outlier_cols = ['Air_Pollution_Index','Water_Pollution_Index','Soil_Pollution_Index',
                'Industrial_Waste (in tons)','CO2_Emissions (in MT)',
                'Energy_Recovered (in GWh)','Plastic_Waste_Produced (in tons)']

fig, axes = plt.subplots(2, 4, figsize=(17, 8))
axes = axes.flatten()

outlier_counts = {}
for i, col in enumerate(outlier_cols):
    data = df[col].dropna()
    Q1, Q3 = data.quantile(0.25), data.quantile(0.75)
    IQR = Q3 - Q1
    n_out = ((data < Q1-1.5*IQR) | (data > Q3+1.5*IQR)).sum()
    outlier_counts[col] = n_out
    ax = axes[i]
    ax.boxplot(data, patch_artist=True, vert=True,
               boxprops=dict(facecolor='#3498db', alpha=0.6),
               medianprops=dict(color='red', lw=2),
               flierprops=dict(markerfacecolor='#e74c3c', marker='o', markersize=5))
    ax.set_title(f"{col.split('(')[0].strip()}\n({n_out} outliers)", fontsize=8, fontweight='bold')

axes[-1].axis('off')
plt.suptitle('PHASE 1 — Outlier Detection (IQR Method)', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
print("Outlier counts:", outlier_counts)
````

````python
#  PHASE 1 | STEP 4b : Winsorization (cap at 5th–95th percentile)
df_clean = df.copy()

for col in outlier_cols:
    lo, hi = df_clean[col].quantile(0.05), df_clean[col].quantile(0.95)
    df_clean[col] = df_clean[col].clip(lower=lo, upper=hi)

print("Outliers capped via Winsorization (5th–95th percentile)")
print(f"   Dataset size preserved: {df_clean.shape[0]} rows × {df_clean.shape[1]} cols")

# Visual check after
fig, axes = plt.subplots(1, 3, figsize=(14, 4))
for ax, col, title in zip(axes,
    ['Air_Pollution_Index','Industrial_Waste (in tons)','CO2_Emissions (in MT)'],
    ['Air Pollution', 'Industrial Waste', 'CO2 Emissions']):
    ax.hist(df[col], bins=20, alpha=0.5, color='#e74c3c', label='Before', edgecolor='white')
    ax.hist(df_clean[col], bins=20, alpha=0.6, color='#2ecc71', label='After', edgecolor='white')
    ax.set_title(f'{title} — Before vs After', fontweight='bold')
    ax.legend()
plt.suptitle('PHASE 1 — Outlier Handling Effect', fontsize=12, fontweight='bold')
plt.tight_layout(); plt.show()
````

##  Cell 5 | Step 5 — Feature Engineering

````python
#  PHASE 1 | STEP 5 : Create new features

# 1. Composite Pollution Score (weighted index across 3 pollution types)
df_clean['Composite_Pollution_Score'] = (
    0.40 * df_clean['Air_Pollution_Index'] +
    0.35 * df_clean['Water_Pollution_Index'] +
    0.25 * df_clean['Soil_Pollution_Index']
)

# 2. Energy Recovery Efficiency (GWh recovered per MWh consumed per capita)
df_clean['Energy_Recovery_Efficiency'] = (
    df_clean['Energy_Recovered (in GWh)'] /
    (df_clean['Energy_Consumption_Per_Capita (in MWh)'] + 1)
)

# 3. Industrial Waste Per Capita
df_clean['Waste_Per_Capita'] = (
    df_clean['Industrial_Waste (in tons)'] /
    (df_clean['Population (in millions)'] + 0.001)
)

# 4. Carbon Intensity (CO2 per unit GDP — inefficiency metric)
df_clean['Carbon_Intensity'] = (
    df_clean['CO2_Emissions (in MT)'] /
    (df_clean['GDP_Per_Capita (in USD)'] + 1)
) * 1e6

# 5. Pollution-Energy Interaction (air pollution × energy consumption)
df_clean['Pollution_Energy_Load'] = (
    df_clean['Composite_Pollution_Score'] *
    df_clean['Energy_Consumption_Per_Capita (in MWh)']
) / 1000

new_feats = ['Composite_Pollution_Score','Energy_Recovery_Efficiency',
             'Waste_Per_Capita','Carbon_Intensity','Pollution_Energy_Load']

print(" 5 new engineered features added:")
for f in new_feats:
    print(f"   {f:<35} mean={df_clean[f].mean():.3f}  std={df_clean[f].std():.3f}")

# Visualize engineered features
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for ax, col, color in zip(axes,
    ['Composite_Pollution_Score','Energy_Recovery_Efficiency','Carbon_Intensity'],
    ['#9b59b6','#27ae60','#e67e22']):
    ax.hist(df_clean[col], bins=22, color=color, edgecolor='white', alpha=0.85)
    ax.set_title(col, fontsize=9, fontweight='bold')
    ax.set_xlabel(f'mean={df_clean[col].mean():.2f}')
plt.suptitle('PHASE 1 — Key Engineered Features', fontsize=12, fontweight='bold')
plt.tight_layout(); plt.show()
````

##  Cell 7 | Step 7 — K-Means: Elbow Method & Optimal K

````python
#  PHASE 2 | STEP 7 : Elbow + Silhouette + DB Index
inertias, silhouettes, db_scores, ch_scores = [], [], [], []
K_range = range(2, 11)

for k in K_range:
    km = KMeans(n_clusters=k, init='k-means++', n_init=15, random_state=42)
    labels = km.fit_predict(X_cluster)
    inertias.append(km.inertia_)
    silhouettes.append(silhouette_score(X_cluster, labels))
    db_scores.append(davies_bouldin_score(X_cluster, labels))
    ch_scores.append(calinski_harabasz_score(X_cluster, labels))

# Elbow + 3 validation metrics
fig, axes = plt.subplots(1, 4, figsize=(18, 4))
fig.suptitle('PHASE 2 — K-Means: Optimal K Selection', fontsize=13, fontweight='bold')

# Elbow
ax = axes[0]
ax.plot(K_range, inertias, marker='o', lw=2.5, color='#3498db', markersize=8)
ax.set_title('Elbow Method\n(Inertia / WCSS)', fontweight='bold')
ax.set_xlabel('K'); ax.set_ylabel('Inertia')
# Highlight elbow
diffs = np.diff(inertias)
diffs2 = np.diff(diffs)
elbow_k = K_range[np.argmax(diffs2) + 2]
ax.axvline(elbow_k, color='red', linestyle='--', lw=2, label=f'Elbow K={elbow_k}')
ax.legend()

# Silhouette
ax = axes[1]
best_sil_k = K_range[np.argmax(silhouettes)]
ax.plot(K_range, silhouettes, marker='s', lw=2.5, color='#27ae60', markersize=8)
ax.axvline(best_sil_k, color='red', linestyle='--', lw=2, label=f'Best K={best_sil_k}')
ax.set_title('Silhouette Score\n(Higher = Better)', fontweight='bold')
ax.set_xlabel('K'); ax.set_ylabel('Silhouette Score')
ax.legend()

# DB Index
ax = axes[2]
best_db_k = K_range[np.argmin(db_scores)]
ax.plot(K_range, db_scores, marker='^', lw=2.5, color='#e74c3c', markersize=8)
ax.axvline(best_db_k, color='red', linestyle='--', lw=2, label=f'Best K={best_db_k}')
ax.set_title('Davies-Bouldin Index\n(Lower = Better)', fontweight='bold')
ax.set_xlabel('K'); ax.set_ylabel('DB Index')
ax.legend()

# CH Score
ax = axes[3]
best_ch_k = K_range[np.argmax(ch_scores)]
ax.plot(K_range, ch_scores, marker='D', lw=2.5, color='#9b59b6', markersize=8)
ax.axvline(best_ch_k, color='red', linestyle='--', lw=2, label=f'Best K={best_ch_k}')
ax.set_title('Calinski-Harabasz\n(Higher = Better)', fontweight='bold')
ax.set_xlabel('K'); ax.set_ylabel('CH Score')
ax.legend()

plt.tight_layout(); plt.show()

# Vote for optimal K
votes = [elbow_k, best_sil_k, best_db_k, best_ch_k]
optimal_k = max(set(votes), key=votes.count)
print(f"Metric votes → Elbow:{elbow_k} | Silhouette:{best_sil_k} | DB:{best_db_k} | CH:{best_ch_k}")
print(f" Optimal K selected: {optimal_k}")
````

##  Cell 8 | Step 8 — K-Means: Fit & Cluster Visualization

````python
#  PHASE 2 | STEP 8 : Fit K-Means with optimal K
kmeans = KMeans(n_clusters=optimal_k, init='k-means++',
                n_init=20, max_iter=500, random_state=42)
df_clean['KMeans_Cluster'] = kmeans.fit_predict(X_cluster)

# Silhouette for final model
final_sil = silhouette_score(X_cluster, df_clean['KMeans_Cluster'])
final_db  = davies_bouldin_score(X_cluster, df_clean['KMeans_Cluster'])
print(f" K-Means fitted with K={optimal_k}")
print(f"   Silhouette Score : {final_sil:.4f}  (closer to 1.0 = better)")
print(f"   Davies-Bouldin   : {final_db:.4f}  (lower = better)")
print(f"\n   Cluster sizes:")
print(df_clean['KMeans_Cluster'].value_counts().sort_index().rename(lambda x: f'Cluster {x}'))
````

````python
#  PHASE 2 | STEP 8b : PCA 2D visualization of clusters
from sklearn.decomposition import PCA

pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_cluster)
explained = pca.explained_variance_ratio_

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('PHASE 2 — K-Means Cluster Visualization (PCA 2D)', fontsize=13, fontweight='bold')

cluster_palette = plt.cm.tab10(np.linspace(0, 1, optimal_k))

ax = axes[0]
for c in range(optimal_k):
    mask = df_clean['KMeans_Cluster'] == c
    ax.scatter(X_pca[mask, 0], X_pca[mask, 1],
               c=[cluster_palette[c]], s=60, alpha=0.75,
               edgecolors='white', lw=0.5, label=f'Cluster {c}')
# Plot centroids
centers_pca = pca.transform(kmeans.cluster_centers_)
ax.scatter(centers_pca[:, 0], centers_pca[:, 1],
           c='black', s=220, marker='X', zorder=5, label='Centroids')
ax.set_xlabel(f'PC1 ({explained[0]*100:.1f}% variance)')
ax.set_ylabel(f'PC2 ({explained[1]*100:.1f}% variance)')
ax.set_title('K-Means Clusters (PCA Projection)', fontweight='bold')
ax.legend(loc='best', fontsize=9)

# Air Pollution vs CO2 colored by cluster
ax = axes[1]
for c in range(optimal_k):
    mask = df_clean['KMeans_Cluster'] == c
    ax.scatter(df_clean[mask]['Air_Pollution_Index'],
               df_clean[mask]['CO2_Emissions (in MT)'],
               c=[cluster_palette[c]], s=60, alpha=0.75,
               edgecolors='white', lw=0.5, label=f'Cluster {c}')
ax.set_xlabel('Air Pollution Index')
ax.set_ylabel('CO2 Emissions (MT)')
ax.set_title('Air Pollution vs CO2 by Cluster', fontweight='bold')
ax.legend(loc='best', fontsize=9)

plt.tight_layout(); plt.show()
````

### Design Decisions

| Decision | Choice | Justification |
|----------|--------|---------------|
| Task type | Regression | Target is a continuous GWh value |
| Loss function | MSE | Standard for regression; penalises large errors more |
| Optimizer | Adam | Adaptive learning rate; robust to sparse gradients |
| Regularisation | Dropout + BatchNorm (baseline); L2 + both (tuned) | Prevents overfitting on small n=200 dataset |

| Seeds | `np.random.seed(42)`, `tf.random.set_seed(42)` | Full reproducibility documented in title cell |


##  Cell 13 | Step 13 — Neural Network: Baseline Model

````python
#  PHASE 3 | STEP 13 : Baseline Feedforward Neural Network
n_features = X_train_nn.shape[1]

def build_baseline_nn(n_in):
    model = keras.Sequential([
        layers.Input(shape=(n_in,)),
        layers.Dense(64,  activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.2),
        layers.Dense(32,  activation='relu'),
        layers.Dropout(0.1),
        layers.Dense(16,  activation='relu'),
        layers.Dense(1)
    ], name='Baseline_NN')

    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=0.001),
        loss='mse',
        metrics=['mae']
    )
    return model

baseline_model = build_baseline_nn(n_features)
baseline_model.summary()
````

##  Cell 16 | Step 16 — Neural Network: Hyperparameter Tuning

````python
#  PHASE 3 | STEP 16 : Tuned Neural Network
# Tuning: deeper architecture, L2 regularization, adjusted LR

def build_tuned_nn(n_in, lr=0.0005):
    model = keras.Sequential([
        layers.Input(shape=(n_in,)),

        layers.Dense(128, activation='relu',
                     kernel_regularizer=regularizers.l2(1e-4)),
        layers.BatchNormalization(),
        layers.Dropout(0.3),

        layers.Dense(64, activation='relu',
                     kernel_regularizer=regularizers.l2(1e-4)),
        layers.BatchNormalization(),
        layers.Dropout(0.2),

        layers.Dense(32, activation='relu'),
        layers.Dropout(0.1),

        layers.Dense(16, activation='relu'),
        layers.Dense(1)
    ], name='Tuned_NN')

    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=lr),
        loss='mse', metrics=['mae']
    )
    return model

tuned_model = build_tuned_nn(n_features, lr=0.0005)
tuned_model.summary()
````

## Cell 18 | Step 18 — Cross-Validation: ML Models + Neural Network

### Why cross-validation is essential here
A single 80/20 train-test split on **n=200** produces unstable metric estimates — variance in R² of ±0.05 or more depending on the random split. K-fold cross-validation (k=5) mitigates this by:
1. Using every data point for both training and validation across folds
2. Averaging metrics across 5 non-overlapping test sets
3. Reporting standard deviation — quantifying how stable performance actually is

**For the Neural Network:** The model is retrained from scratch for each fold (identical architecture, same seeds + fold offset), proving that strong performance is not an artefact of a lucky single split.

**Interpretation guide:**
- CV R² std < 0.05 → stable, reliable model
- CV R² std > 0.10 → high sensitivity to training data; treat results cautiously
- Compare CV R² to test-set R² — large gap may indicate overfitting to the test partition


````python
# ══════════════════════════════════════════════════════════════════
# CELL 18 | NEW — 5-Fold Cross-Validation (ML + Neural Network)
# Addresses mentor feedback: model validation using cross-validation
# Random seeds: np.random.seed(42) + tf seed set per fold for full
# reproducibility — documented in title cell.
# ══════════════════════════════════════════════════════════════════

from sklearn.model_selection import KFold, cross_validate
from sklearn.metrics import make_scorer

kf = KFold(n_splits=5, shuffle=True, random_state=42)

# ── 1. 5-Fold CV for scikit-learn regression models ──────────────
cv_ml_models = {
    'Linear Regression':  LinearRegression(),
    'Ridge (α=1.0)':      Ridge(alpha=1.0),
    'Lasso (α=0.1)':      Lasso(alpha=0.1, max_iter=5000),
    'Gradient Boosting':  GradientBoostingRegressor(n_estimators=100, random_state=42),
    'Random Forest':      RandomForestRegressor(n_estimators=100, random_state=42),
}

rmse_scorer = make_scorer(lambda y, p: -np.sqrt(mean_squared_error(y, p)))
scoring = {'r2': 'r2', 'rmse': rmse_scorer, 'mae': 'neg_mean_absolute_error'}

cv_summary = {}
print(f"{'Model':<25} {'CV R²':>8} {'±Std':>7} {'CV RMSE':>10} {'CV MAE':>10}")
print("—" * 65)

for name, model in cv_ml_models.items():
    res = cross_validate(model, X_nn, y_nn, cv=kf, scoring=scoring)
    r2_mean  = res['test_r2'].mean()
    r2_std   = res['test_r2'].std()
    rmse_m   = -res['test_rmse'].mean()
    mae_m    = -res['test_mae'].mean()
    cv_summary[name] = {'r2': r2_mean, 'r2_std': r2_std, 'rmse': rmse_m, 'mae': mae_m}
    print(f"{name:<25} {r2_mean:>8.4f} {r2_std:>7.4f} {rmse_m:>10.4f} {mae_m:>10.4f}")

# ── 2. 5-Fold CV for Tuned Neural Network (manual loop) ──────────
print("\n  Running Neural Network 5-Fold CV...")
nn_r2, nn_rmse, nn_mae = [], [], []

for fold_i, (tr_idx, vl_idx) in enumerate(kf.split(X_nn)):
    tf.random.set_seed(42 + fold_i)          # per-fold seed: reproducible
    X_tr, X_vl = X_nn[tr_idx], X_nn[vl_idx]
    y_tr, y_vl = y_nn[tr_idx], y_nn[vl_idx]

    fold_model = build_tuned_nn(X_nn.shape[1], lr=0.0005)
    fold_model.fit(X_tr, y_tr,
        validation_data=(X_vl, y_vl),
        epochs=200, batch_size=16,
        callbacks=[
            callbacks.EarlyStopping(monitor='val_loss', patience=25,
                                    restore_best_weights=True, verbose=0),
            callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5,
                                        patience=10, min_lr=1e-7, verbose=0)
        ], verbose=0)

    yp = fold_model.predict(X_vl, verbose=0).flatten()
    nn_r2.append(r2_score(y_vl, yp))
    nn_rmse.append(np.sqrt(mean_squared_error(y_vl, yp)))
    nn_mae.append(mean_absolute_error(y_vl, yp))
    print(f"     Fold {fold_i+1}: R²={nn_r2[-1]:.4f}  RMSE={nn_rmse[-1]:.4f}")

nn_r2_arr = np.array(nn_r2)
cv_summary['Tuned NN'] = {
    'r2': nn_r2_arr.mean(), 'r2_std': nn_r2_arr.std(),
    'rmse': np.mean(nn_rmse), 'mae': np.mean(nn_mae)
}
print(f"\n{'Tuned NN (5-Fold)':<25} {nn_r2_arr.mean():>8.4f} {nn_r2_arr.std():>7.4f} "
      f"{np.mean(nn_rmse):>10.4f} {np.mean(nn_mae):>10.4f}")

# ── 3. Visualise CV results ───────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle('5-Fold Cross-Validation Results — All Models', fontsize=13, fontweight='bold')

all_names = list(cv_summary.keys())
r2_means  = [cv_summary[m]['r2']     for m in all_names]
r2_stds   = [cv_summary[m]['r2_std'] for m in all_names]
best_idx  = int(np.argmax(r2_means))
bar_cols  = ['#e74c3c' if 'NN' in m else '#3498db' for m in all_names]
bar_cols[best_idx] = '#2ecc71'

ax = axes[0]
bars = ax.bar(range(len(all_names)), r2_means, color=bar_cols, alpha=0.85,
              edgecolor='white', yerr=r2_stds, capsize=6,
              error_kw={'ecolor': 'black', 'lw': 2})
ax.set_xticks(range(len(all_names)))
ax.set_xticklabels(all_names, rotation=30, ha='right', fontsize=9)
ax.set_ylabel('CV R² (Mean ± 1 Std)')
ax.set_title('Model Comparison — CV R²\n(Error bars = ±1 Std across folds)', fontweight='bold')
ax.set_ylim(max(0, min(r2_means) - 0.15), 1.1)
for bar, val, std in zip(bars, r2_means, r2_stds):
    ax.text(bar.get_x() + bar.get_width()/2, val + std + 0.015,
            f'{val:.3f}', ha='center', fontsize=9, fontweight='bold')

ax = axes[1]
fold_lbls = [f'Fold {i+1}' for i in range(5)]
ax.plot(fold_lbls, nn_r2, marker='o', lw=2.5, color='#e74c3c',
        markersize=10, zorder=5, label='NN R² per fold')
ax.axhline(nn_r2_arr.mean(), color='#e74c3c', lw=2, linestyle='--',
           label=f'Mean = {nn_r2_arr.mean():.4f}')
ax.fill_between(range(5),
                nn_r2_arr.mean() - nn_r2_arr.std(),
                nn_r2_arr.mean() + nn_r2_arr.std(),
                alpha=0.18, color='#e74c3c', label=f'±1 Std = {nn_r2_arr.std():.4f}')
ax.set_ylabel('R²')
ax.set_title('Neural Network — R² per CV Fold\n(Stability check across data subsets)', fontweight='bold')
ax.legend(fontsize=9)
ax.set_ylim(max(0, nn_r2_arr.mean() - 0.3), 1.05)

plt.tight_layout()
plt.show()

print(f"\n=== CV INTERPRETATION ===")
print(f"  Best model by CV R²: {all_names[best_idx]} ({r2_means[best_idx]:.4f})")
print(f"  Tuned NN CV R²:      {nn_r2_arr.mean():.4f} ± {nn_r2_arr.std():.4f}")
stability = 'stable' if nn_r2_arr.std() < 0.05 else 'moderate variance'
print(f"  NN stability:        {stability} across folds (std={nn_r2_arr.std():.4f})")
````

##  Cell 19 | Step 19 — Feature Importance & Explainability Analysis

### Why explainability matters
A model that predicts well but cannot explain *why* is not policy-actionable. This cell provides three complementary explainability approaches:

1. **Gradient Boosting feature importances** — tree-based Gini impurity reduction; fast, interpretable
2. **Random Forest feature importances** — averaged across 100 trees; more stable than single DT
3. **Permutation importance (model-agnostic)** — shuffles each feature and measures R² drop; works on *any* model including the neural network, making it the most honest measure of feature influence

**Policy interpretation:** Features with the highest permutation importance are the variables that most drive predicted energy recovery. Targeting these in policy interventions yields the highest expected returns.

**Partial dependence plots (PDPs):** Show the *marginal* effect of each top feature on predicted energy recovery, holding all other features at their mean. This directly answers: *"If we improve Renewable_Energy % by X points, by how much does predicted energy recovery increase?"*


````python
# ══════════════════════════════════════════════════════════════════
# CELL 19 | NEW — Feature Importance & Explainability
# Addresses mentor feedback: feature importance / SHAP-style analysis
# Uses permutation importance (model-agnostic, sklearn built-in),
# GB & RF tree importances, and partial dependence plots.
# ══════════════════════════════════════════════════════════════════

# ── 1. Gradient Boosting & Random Forest importances ─────────────
gb_model = GradientBoostingRegressor(n_estimators=200, max_depth=4,
                                      learning_rate=0.05, random_state=42)
rf_model = RandomForestRegressor(n_estimators=200, random_state=42)
gb_model.fit(X_train_nn, y_train_nn)
rf_model.fit(X_train_nn, y_train_nn)

gb_imp = pd.Series(gb_model.feature_importances_, index=nn_features).sort_values(ascending=False)
rf_imp = pd.Series(rf_model.feature_importances_, index=nn_features).sort_values(ascending=False)

print("=== GRADIENT BOOSTING Feature Importances (top 10) ===")
for feat, val in gb_imp.head(10).items():
    bar = '█' * int(val * 300)
    print(f"  {feat:<40} {val:.4f}  {bar}")

print("\n=== RANDOM FOREST Feature Importances (top 10) ===")
for feat, val in rf_imp.head(10).items():
    bar = '█' * int(val * 300)
    print(f"  {feat:<40} {val:.4f}  {bar}")

# ── 2. Permutation Importance on Tuned NN (model-agnostic) ───────
print("\nComputing permutation importance on Tuned NN (20 repeats)...")
np.random.seed(42)

def r2_score_manual(y_true, y_pred):
    ss_res = np.sum((y_true - y_pred) ** 2)
    ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
    return 1 - ss_res / ss_tot

baseline_r2 = r2_score_manual(y_test_nn,
                               tuned_model.predict(X_test_nn, verbose=0).flatten())
perm_drops = []
for fi in range(len(nn_features)):
    drops = []
    for _ in range(20):
        X_p = X_test_nn.copy()
        X_p[:, fi] = X_p[np.random.permutation(len(X_p)), fi]
        r2_p = r2_score_manual(y_test_nn,
                                tuned_model.predict(X_p, verbose=0).flatten())
        drops.append(baseline_r2 - r2_p)
    perm_drops.append(np.mean(drops))

perm_df = pd.DataFrame({'Feature': nn_features, 'PermImp': perm_drops})
perm_df = perm_df.sort_values('PermImp', ascending=False).reset_index(drop=True)

print("\n=== PERMUTATION IMPORTANCE (Tuned NN) — mean R² drop ===")
for _, row in perm_df.iterrows():
    bar = '█' * max(0, int(row['PermImp'] * 300))
    print(f"  {row['Feature']:<40} {row['PermImp']:+.4f}  {bar}")

# ── 3. Visualisation
fig, axes = plt.subplots(1, 3, figsize=(20, 7))
fig.suptitle('Feature Importance & Explainability Analysis', fontsize=14, fontweight='bold')

# Panel A — GB importances
ax = axes[0]
top_gb = gb_imp.head(12).sort_values()
cols_gb = ['#e74c3c' if v > top_gb.median() else '#aed6f1' for v in top_gb]
ax.barh(top_gb.index, top_gb.values, color=cols_gb, alpha=0.88, edgecolor='white')
ax.set_title('''Gradient Boosting
Feature Importances
(Gini impurity reduction)''',
             fontweight='bold', fontsize=9)
ax.set_xlabel('Importance')

# Panel B — RF importances
ax = axes[1]
top_rf = rf_imp.head(12).sort_values()
cols_rf = ['#3498db' if v > top_rf.median() else '#aed6f1' for v in top_rf]
ax.barh(top_rf.index, top_rf.values, color=cols_rf, alpha=0.88, edgecolor='white')
ax.set_title('''Random Forest
Feature Importances
(Mean decrease in impurity)''',
             fontweight='bold', fontsize=9)
ax.set_xlabel('Importance')

# Panel C — Permutation importance (NN)
ax = axes[2]
top_perm = perm_df.head(12).sort_values('PermImp')
cols_perm = ['#e67e22' if v > 0 else '#aaaaaa' for v in top_perm['PermImp']]
ax.barh(top_perm['Feature'], top_perm['PermImp'],
        color=cols_perm, alpha=0.88, edgecolor='white')
ax.axvline(0, color='black', lw=1.5, linestyle='--')
ax.set_title('''Permutation Importance
(Tuned Neural Network)
Mean R² drop when shuffled''',
             fontweight='bold', fontsize=9)
ax.set_xlabel('R² drop (higher = more important)')

plt.tight_layout()
plt.show()

# ── 4. Partial Dependence Plots — top 3 features (Tuned NN) ──────
top3 = perm_df['Feature'].head(3).tolist()
top3_idx = [nn_features.index(f) for f in top3]

fig, axes = plt.subplots(1, 3, figsize=(17, 5))
fig.suptitle('''Partial Dependence Plots — Top 3 Features (Tuned NN)
Marginal effect on predicted Energy Recovery (GWh)''',
             fontsize=13, fontweight='bold')

for ax, feat, fi in zip(axes, top3, top3_idx):
    grid = np.linspace(X_test_nn[:, fi].min(), X_test_nn[:, fi].max(), 60)
    X_pdp = np.tile(X_nn.mean(axis=0), (60, 1))
    X_pdp[:, fi] = grid

    orig_std  = df_clean[feat].std()  if feat in df_clean.columns else 1
    orig_mean = df_clean[feat].mean() if feat in df_clean.columns else 0
    x_orig = grid * orig_std + orig_mean
    y_pdp  = tuned_model.predict(X_pdp, verbose=0).flatten()

    ax.plot(x_orig, y_pdp, lw=2.5, color='#e74c3c')
    ax.fill_between(x_orig, y_pdp, alpha=0.12, color='#e74c3c')
    ax.axvline(orig_mean, color='gray', lw=1.5, linestyle='--',
               label=f'Mean = {orig_mean:.1f}')
    lbl = feat.split('(')[0].strip().replace('_', ' ')
    ax.set_xlabel(lbl, fontsize=9)
    ax.set_ylabel('Predicted Energy Recovered (GWh)')
    ax.set_title(f'''PDP: {lbl}
(Marginal effect, all else = mean)''',
                 fontweight='bold', fontsize=9)
    ax.legend(fontsize=8)

plt.tight_layout()
plt.show()

# ── 5. Rank comparison: GB vs RF vs NN Permutation ───────────────
fig, ax = plt.subplots(figsize=(12, 5))
fig.suptitle('Feature Importance Rank Comparison Across Methods',
             fontsize=13, fontweight='bold')

top_feats_union = perm_df['Feature'].head(8).tolist()
gb_rank   = {f: i+1 for i, f in enumerate(gb_imp.index)}
rf_rank   = {f: i+1 for i, f in enumerate(rf_imp.index)}
perm_rank = {row['Feature']: i+1 for i, row in perm_df.iterrows()}

x = np.arange(len(top_feats_union))
ax.plot(x, [perm_rank.get(f, 14) for f in top_feats_union],
        'o-', lw=2.5, color='#e67e22', markersize=9, label='Permutation Imp (NN)')
ax.plot(x, [gb_rank.get(f, 14) for f in top_feats_union],
        's-', lw=2.5, color='#3498db', markersize=9, label='Gradient Boosting')
ax.plot(x, [rf_rank.get(f, 14) for f in top_feats_union],
        '^-', lw=2.5, color='#2ecc71', markersize=9, label='Random Forest')
ax.set_xticks(x)
ax.set_xticklabels([f.split('(')[0].strip()[:18] for f in top_feats_union],
                    rotation=30, ha='right', fontsize=9)
ax.set_ylabel('Rank (1 = most important)')
ax.invert_yaxis()
ax.set_title('''Features ranked 1–14 by three methods
Convergence across methods = high confidence''', fontweight='bold')
ax.legend(fontsize=9)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("\n=== EXPLAINABILITY SUMMARY ===")
print(f"  Baseline model R² on test set : {baseline_r2:.4f}")
print(f"  Top 3 features by permutation importance:")
for i, (_, row) in enumerate(perm_df.head(3).iterrows()):
    print(f"    {i+1}. {row['Feature']:<38}  R² drop = {row['PermImp']:+.4f}")
print("\n  Policy implication: Targeting these features in interventions")
print("  yields the highest predicted improvement in energy recovery.")
````

### Section C — Limitations and Honest Caveats

**Correlation ≠ Causation.** The NN identifies correlational patterns from observational data. The policy scenario simulations (e.g., CO2 −20% → energy recovery change) are *model-conditional predictions*, not guaranteed causal outcomes. Factors not captured in the dataset (geopolitical stability, technology access, regulation quality) could dominate real-world outcomes.

**Small dataset (n=200).** Even with cross-validation, 200 records produces metric estimates with non-trivial variance. Results should be validated on larger, independently collected datasets before being used for policy decisions at national scale.

**No geospatial or temporal structure.** The model treats each country-year as an independent observation, ignoring geographic spillover effects (pollution crossing borders) and autoregressive dynamics (this year's pollution affects next year's recovery). A panel data model or LSTM would capture these relationships.
